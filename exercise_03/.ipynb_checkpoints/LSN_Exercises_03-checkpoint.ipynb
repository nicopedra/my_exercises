### Exercise 03.1 <span style="color:red"> Plain vanilla option pricing </span>

Black-Scholes theory assumes that the time evolution of the asset price is stochastic and exhibits geometric Brownian motion (GBM) with constant risk-free interest rate $r$ and volatility $\sigma$.

Black-Scholes <span style="color:blue"> analytic solution </span> for the price at time t<T:
- European Call-option price:
$$ C\left[ S(t),t \right] = S(t) N(d_1) -K e^{-r(T-t)} N(d_2) $$
- European Put-option price:
$$ C\left[ S(t),t \right] = S(t) \left[ N(d_1)-1\right] -K e^{-r(T-t)} \left[ N(d_2)-1\right]$$

where 
$$d_1 = \frac{1}{\sigma\sqrt{T-t}}\left[ \ln \left( \frac{S(t)}{K} \right) + \left( r+\frac{\sigma^2}{2} (T-t) \right)\right]$$

where $d_2 = d_1 - \sigma\sqrt{T-t}$, and where
$$ N(x) = \int_{-\infty}^{x} dz \, e^{-\frac{z^2}{2}} = \frac{1}{2} \left[ 1+ {\rm erf} \left( \frac{x}{\sqrt{2}}\right) \right]$$
is the cumulative of Gaussian $\sim \mathcal{N}(\mu=0,\sigma^2=1)$,
and where
$${\rm erf}(x) = \frac{2}{\sqrt{\pi}}\int_{0}^{x} dz \, e^{-z^2}$$

By using the following parameters: 
- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$

compute at time $t=0$ **via Monte Carlo** the European call-option price, $C[S(0),0]$, and put-option price, $P[S(0),0]$:

1. <span style="color:red"> By sampling **directly** the final asset price $S(T)$</span> for a $GBM(r,\sigma^2)$
2. <span style="color:red"> By sampling the discretized $GBM(r,\sigma^2)$ path of the asset price</span> 
dividing $[0,T]$ in $10^2$ time intervals: $S(0)$, $S(t_1)$, $S(t_2)$, ... up to $S(t_{100}=T=1)$

<p style="border:2px; border-style:solid; border-color:#F5F5F5; padding: 1em; background-color:#F5F5F5">
    Remember that given an asset price $S\sim GBM(\mu,\sigma^2)$, with initial value $S(0)$, then
    $$S(t) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)t+\sigma W(t)\right]$$
    where $W(t)\sim\mathcal{N}(0,t)$.
    Moreover, since the increments of $W$ are independent and normally distributed, the simple recursive procedure for sampling discretized values of $S\sim GBM(\mu,\sigma^2)$ at $t_1 < t_2 < … < t_n$ is:
$$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$
    with $Z_i \sim \mathcal{N}(0,1)$
</p>

### For comparison, Black-Scholes analytic solution:

```python
import scipy
from scipy.special import erf
#from scipy import log, exp, sqrt
from numpy import log, exp, sqrt

# data
S0 = 100.
K = 100.
T = 1.
r = 0.1
sigma = 0.25
t=0

def N(x):
    return 0.5 * (1. + erf(x / sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * sqrt(T)) * (log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * sqrt(T)
    C = S0 * N(d1) - K * exp(-r * T) * N(d2)
    P = S0 *(N(d1) - 1.) - K * exp(-r * T) * (N(d2)-1.)
    return C, P

if __name__ == "__main__":
    call, put = black_scholes(S0, K, T, r, sigma)
    print(" == BLACK-SCHOLES ==")
    print("call: ", call)
    print("put: ", put)
```

### DIRECT WAY

Per il metodo diretto è stata implementata una funzione chiamata direttamente nel main che prende come argomenti:  
- generatore di numeri casuali  
- numero totale di iterazioni
- numero di blocchi per il blocking
- i parametri S0,K,T,r,sigma
- i valori di C e P ottenuti con la soluzione analitica (anch'essa implementata nel codice in $\texttt{c++}$)
  
Nel metodo diretto ad ogni step all'interno del blocco vengono accumulate le quantità:  
1. $\exp(-rT) \max (0.,value-K)$ per C
2. $\exp(-rT) \max (0.,K-value)$ per P  
  
dove $value= S0 \exp{ (r-\frac{\sigma^2}{2})T - \sigma x \sqrt{T} }$, e x è un numero casuale generato secondo una distribuzione gaussiana centrata in 0 e con varianza 1

### DISCRETIZED WAY

Per il metodo discreto (cioè in cui viene discretizzato il tempo di delivery T), viene chiamata una funziona analoga alla precedente ma con un parametro in più:    
- $n_{step}$, il numero di step in cui dividere l'intervallo $[0,T]$  
  
Successivamente quello che viene fatto è questo:  
come prima ad ogni step all'interno di un singolo blocco vengono accumulate le quantità:  
1. $\exp(-rT) \max (0.,value-K)$ per C
2. $\exp(-rT) \max (0.,K-value)$ per P  
  
dove ora però ora $value$ viene ricavato compiendo un ciclo for sul numero di $n_{step}$ e accumulando il prodotto di $value *= S0 \exp{ (r-\frac{\sigma^2}{2})\frac{T}{n_{step}} - \sigma x \sqrt{\frac{T}{n_{step}}} }$ ad ogni iterazione, e x è sempre un numero casuale generato secondo una distribuzione gaussiana centrata in 0 e con varianza 1

## RESULTS from MC simulations

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
matplotlib.rcParams['text.usetex'] = True

M=10000
N=100
#L=int(M/N)
x = np.arange(N)
y = np.zeros(N)
#x*=L
name = "directC","directP","discreteC","discreteP"
title = r"\textbf{direct sampling for C}",r"\textbf{direct sampling for P}",r"\textbf{discretized sampling for C}",r"\textbf{discretized sampling for P}"
fig,axs = plt.subplots(nrows=4,ncols=1,figsize=(15,25) )

for i in range(4):
    sum_prog,err_prog = np.loadtxt(name[i] + ".txt",usecols=(0,1), delimiter=' ', unpack='true')
    axs[i].errorbar(x,sum_prog,yerr=err_prog,color='b',ecolor='r')
    axs[i].plot(x,y,color='black')
    if i%2==0:
        axs[i].set_ylabel(name[i]+f"$-${round(call,2)}")
        axs[i].legend(["analytical call","MC call"])
    if i%2==1:
        axs[i].set_ylabel(name[i]+f"$-${round(put,2)}")
        axs[i].legend(["analytical put","MC put"])
    #axs[i].set_xlabel("throwns")
    axs[i].set_xlabel("number of blocks")
    axs[i].set_title(title[i])
    

plt.show()
```
