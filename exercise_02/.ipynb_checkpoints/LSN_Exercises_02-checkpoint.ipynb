### Exercise 02.1
- Compute the following 1D integral via Monte Carlo 
$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$

1. sampling a uniform distribution in $[0,1]$

La prima parte dell'esercitazione consiste nel valutare l'integrale sopra scritto campionando i punti di integrazione in modo uniforme e successivamente (nella seconda parte) utilizzando il metodo dell'*importance sampling*  
  
1. Integrare utilizzando il primo metodo significa semplicemente generare in modo uniforme i valori di $x$ (usando il generatore di numeri casuali usato nell'esercizio 01) nell'intervallo di integrazione $[0,1]$ (in generale $[min,max]$). Dopo aver generato un valore di $x$ occorre valutare la funzione integranda con quel valore e accumulare i vari valori ottenuti (per ogni $x$ generato) in una variabile, ad esempio $Uniform\_I$.  
A questo punto il risultato dell'integrale è semplicemente $ \frac{Uniform\_I}{total\_throwns} \cdot (max-min)$, dove $total\_throwns$ corrisponde al numero totale di x generate 
  
Il grafico mostra il risultato ottenuto analizzando i dati usando il metodo blocking. Ho utilizzato 100 blocchi con 100 valori per blocco. Per ogni blocco viene calcolata una stima dell'integrale e poi viene chiamata la funzione già citata nell'esercizio 01 che svolge l'analisi dati.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
matplotlib.rcParams['text.usetex'] = True

M=10000
N=100
L=int(M/N)
x = np.arange(N)
#x*=L
y = np.zeros(N)
sum_prog,err_prog = np.loadtxt("data_uniform_I.txt",usecols=(0,1), delimiter=' ', unpack='true')
plt.errorbar(x,sum_prog,yerr=err_prog,color='b',ecolor='r')
plt.ylabel(r"$<I>-1$")
#plt.xlabel("throwns")
plt.xlabel("number of blocks")
plt.plot(x,y,color='black')

plt.show()
```

2. using importance sampling (i.e. sampling a non-uniform probability in $[0,1]$)

<span style="color:red">Show a picture of both your estimations of $I$</span> and their uncertainties with a large number of *throws* $M$ (e.g. $M\ge 10^4$) as a function of the number of blocks, $N$</font>.

2. Il secondo metodo, dell'importance sampling, è molto più interessante. L'obbiettivo è minimizzare l'errore campionando in modo diverso e intelligente i punti relativi all'integrazione. Ricordando che $I = \int g(x)p(x) dx = \int \frac {g(x)p(x)}{h(x)} h(x) dx$, con $h(x) \geq 0$ e $\int h(x) dx = 1$.  
Si è giunti al medesimo integrale, ma con una interpretazione diversa. Ora l'obbiettivo è campionare secondo $h(x)$ la funzione integranda $\frac {g(x)p(x)}{h(x)}$. Più $h(x)$ è simile alla funzione integranda più la varianza diminuisce, per cui diminuisce l'errore sul calcolo dell'integrale.  
Nel mio caso specifico ho scelto come funzione per l'importance samplig la seguente: $\frac{\pi}{2} \cdot (1-x)$, con la condizione di normalizzazione la distribuzione di probabilità è: $h(x) = 2 (1-x)$. La scelta di questa funzione la si può intedendere mostrandola grafica insieme a $\frac{\pi}{2} \cos(\frac{\pi x}{2})$

```python
x = np.linspace(0,1,100)
y = np.cos(np.pi * x*0.5)
y = np.pi*0.5*y
z = np.pi*0.5 *(1-x)
plt.plot(x,y)
plt.plot(x,z)
plt.title("why choosing a straight line for importance sampling")
plt.legend(["integrand","straight line"])
plt.ylabel("f(x)")
plt.xlabel("x")
plt.show()
```

Per generare un numero distribuito secondo questa distribuzione viene utilizzato il metodo della trasformata:  
- considero una variabile $y$ distribuita in modo uniforme in $[0,1]$ e la pongo uguale alla cumulativa di $h(x)$
- $y = \int_0^x dx h(x) = 2x - x^2$ 
- $x = 1 - \sqrt{1-y}$ (ho scartato la possibilità con il $+$, perchè voglio valori in $[0,1]$
Ora il procedimento è analogo al metodo di uniforme. Vengono generate delle x con questa distribuzione e viene valutata la nuova funzione integranda $\dfrac {\dfrac{\pi}{4} \cos (\frac {\pi}{2})}{1-x}$  
Il procedimento poi è analogo a quello già spiegato sopra  
  
#### considerazioni  
Si può notare dal grafico ottenuto che l'incertezza sul valore dell'integrale è effettivamente diminuita (di un ordine di grandezza) rispetto a quella ottenuta tramite il metodo dell'integrale uniforme. (significa che la scelta della retta è stata una buona scelta)

```python
x = np.arange(N)
#x*=L
y = np.zeros(N)
sum_prog,err_prog = np.loadtxt("data_importance_I.txt",usecols=(0,1), delimiter=' ', unpack='true')
plt.errorbar(x,sum_prog,yerr=err_prog,color='b',ecolor='g')
plt.ylabel(r"$<I>-1$")
#plt.xlabel("throwns")
plt.xlabel("number of blocks")
plt.plot(x,y,color='black')

plt.show()
```


### Exercise 02.2
- <span style="color:red">3D Random Walks (RW) on a cubic lattice and in the continuum:</span> Repeat many times (e.g. say $10^4$) the simulation of a random walk in 3D always starting at the origin

Per prima cosa bisogna far muovere il **random walker** nello spazio tridimensionale:  
- nel caso discreto ho fatto questo: scelgo in modo random (con il solito generatore) un numero intero tra $0$ e $2$, che indica la coordinata nello spazio $3D$ da muovere. Successivamente genero un altro numero casuale in $[0,1]$ e lo confronto con $0.5$, se è minore allora il random walker compie uno step indietro, altrimenti in avanti nella direzione della coordinata estrapolata.
- nel caso continuo : genero in modo random un angolo $\phi$ in $[0,2\pi]$ e un angolo $\theta = \arccos(1-x)$ dove x è un numero distribuito in modo uniforme in $[0,1]$ e mi muovo in questa direzione generata. Il vettore restituito è sempre in coordinate cartesiane, per cui il passo che faccio è $(a\sin(\theta) \cos(\phi) , a\sin(\theta) \sin(\phi),a\cos(\theta))$ 
  
In entrambi i casi il passo $a$ è stato posto uguale a $1$ e sono sempre partiti dall'origine.
  
A questo definito il modo in cui muovere il random walker si eseguono $10^4$ simulazioni identiche:
- partendo dall'origine si muove il random walker di $10^2$ passi
- ad ogni passo si salva $\sqrt{ |\vec{r}_N|^2 _{RW}}$, ottenendo quindi alla fine $10^4$ valori per ogni passo svolto
- finita la simulazione per ogni passo viene svolta l'analisi dati con il metodo blocking. In questo caso non vengono mostrati i risultati relativi all'analisi (sarebbero 100 grafici da mostrare). Dall'analisi viene ricavato il valore finale e il suo errore (di $\sqrt{ |\vec{r}_N|^2 _{RW}}$), per poi plottarlo insieme agli altri dati ottenuti e mostrare quindi l'andamento di $\sqrt{ |\vec{r}_N|^2 _{RW}}$ in funzione degli step
- l'ultima parte è semplicemente plottare il risultato facendo un fit con una funzione tipo $k \sqrt{x}$, dove mi aspetto che k sia molto simile a $1$ in questo caso proprio perchè lo step del RW è stato posto uguale a $1$ (a)
  
I primi tre punti sono tutti contenuti in una funzione che viene chiamata nel main e che prende come argomenti:  

- generatore di numeri casuali = rnd 
- numero di simulazioni da eseguire = $10^4$     
- numeri di blocchi per il blocking = $10^2$    
- numero di step per il random walk = $10^2$   
- file in cui salvare i dati ottenuti   
- parametro intero che decide se eseguire un RW discreto (0) o continuo (1) 


1. on a cubic lattice with lattice constant $a=1$; at each discrete time the walker makes a forward or backward step of length equal to $a$ in one of the 3 principal directions of the lattice: $x$, $y$ or $z$

```python
from scipy.optimize import curve_fit

def f(x,k):  # defining the fitting function
    return k * np.sqrt(x)

x=np.arange(N)
sum_prog,err_prog = np.loadtxt("data_RW_discr.txt",usecols=(0,1), delimiter=' ', unpack='true')

fig,axs = plt.subplots(nrows=1,ncols=2,figsize=(17, 5))
### without fitting ##############################
axs[0].errorbar(x,sum_prog,yerr=err_prog,color='yellow',ecolor='darkred',alpha=0.8)
axs[0].set_ylabel(r"$\sqrt{<r^2>}$")
axs[0].set_xlabel("N\_step")
axs[0].set_title(r'without fitting')
#### fitting with K*sqrt(x) ######################
axs[1].errorbar(x,sum_prog,yerr=err_prog,color='yellow',ecolor='darkred',alpha=0.8)
axs[1].set_ylabel(r"$\sqrt{<r^2>}$")
axs[1].set_xlabel("N\_step")
axs[1].set_title(r'Fitting a discrete\_RW' + ' $\sqrt{<r^2>}$ with $ k \sqrt{N}$')
p_opt, p_cov = curve_fit(f, x, sum_prog)
y_fit = f(x,p_opt[0])
axs[1].plot(x,y_fit) # plotting fitted function
print("optimized parameters k =" + f" {p_opt}")
print("parameters uncertainty ="f" {np.sqrt(np.diagonal(p_cov))}")

plt.show()
```

2. in the continuum; at each discrete time the walker makes a step of length equal to $a(=1)$ along a **random direction** obtained by sampling **uniformly** the solid angle: $\theta \in [0,\pi]$ and $\phi \in [0,2\pi]$

<span style="color:red">Show a picture of $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ for both RWs</span>, with their statistical uncertainties, as function of the step $i\in [0,10^2]$.

```python
from scipy.optimize import curve_fit

def f(x,k):  # defining the fitting function
    return k * np.sqrt(x)

x=np.arange(N)
sum_prog,err_prog = np.loadtxt("data_RW_cont.txt",usecols=(0,1), delimiter=' ', unpack='true')

fig,axs = plt.subplots(nrows=1,ncols=2,figsize=(17, 5))
### without fitting ##############################
axs[0].errorbar(x,sum_prog,yerr=err_prog,color='purple',ecolor='red',alpha=0.8)
axs[0].set_ylabel(r"$\sqrt{<r^2>}$")
axs[0].set_xlabel("N\_step")
axs[0].set_title(r'without fitting')
#### fitting with K*sqrt(x) ######################
axs[1].errorbar(x,sum_prog,yerr=err_prog,color='purple',ecolor='red',alpha=0.8)
axs[1].set_ylabel(r"$\sqrt{<r^2>}$")
axs[1].set_xlabel("N\_step")
axs[1].set_title(r'Fitting a continue\_RW' + ' $\sqrt{<r^2>}$ with $ k \sqrt{N}$')
p_opt, p_cov = curve_fit(f, x, sum_prog)
y_fit = f(x,p_opt[0])
axs[1].plot(x,y_fit) # plotting fitted function
print("optimized parameters k =" + f" {p_opt}")
print("parameters uncertainty ="f" {np.sqrt(np.diagonal(p_cov))}")

plt.show()
```

Note that you could try to fit both results with a function like $f(N)=k\sqrt{N}$. Do your results indicate a diffusive behavior?

- yes they do. they are RW walk processes with equal probability for stepforward or step backward. So we can find that the probability density to find our RW in a point x at time t follows the diffusion equation: 
- $\frac{\partial p(x,t)}{\partial t} = D \frac{\partial^2 p(x,t)}{\partial x^2}$ 
- if our initial condition is $\delta (x)$ at time t=0 we can find $ p(x,t) = \frac{1}{\sqrt{4 \pi D t}} \exp{\frac{-x^2}{4 D t}} $
- we already know it because of the CLT, and if we try to find $<x^2>$ we find that it is equal to $\sigma^2=N a$ with $N$ the number of step
- note: this is true for all of the three dimension. The movement along one direction is not correleted to the others  

#### COMMENTI FINALI
Come mostrato nei grafici l'andamento della curva $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ fitta perfettamente (entro gli errori) con la curva $k\sqrt{N}$. (e come mi aspettavo il valore di $k$ risulta essere compatibile con il valore del passo RW)    
Questo cosa ci dice?  
Innazitutto ricordiamo che abbiamo a che fare con un processo random walk, processo di Markow, per cui sappiamo ricavare che la distribuzione di probabilità evolve secondo l'equazione di diffusione $\dfrac{\partial p(x,t)}{\partial t} = D \dfrac{\partial^2 p(x,t)}{\partial x^2}$. Se risolviamo per il caso di condizione iniziale $p(x,0) = \delta(x)$ (che significa proprio iniziare a diffondersi partendo dall'origine, esattamente il nostro caso) ricaviamo come soluzione $ p(x,t) = \frac{1}{\sqrt{4 \pi D t}} \exp{\frac{-x^2}{4 D t}} $.  
il momento secondo della distribuzione (la varianza, quello che stiamo graficando e fittando con $k \sqrt{N}$) risulta proprio essere uguale a $\sigma^2 = a^2 N$  
Questi risultati posso essere ottenuti anche considerando il CLT.
