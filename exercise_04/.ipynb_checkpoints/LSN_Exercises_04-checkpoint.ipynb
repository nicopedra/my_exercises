#### Input()
i parametri necessari all'esecuzione del codice sono contenuti nel file input.dat che viene letto tramite la funzinoe input(). Un ulteriore parametro tuttavia viene dato da terminale, il parametro **restart**, che può valere o $0$ o $1$:  
- $0$ -> la configurazione al tempo $t_0$ della simulazione corrisponde ad una configurazione cristallina fcc, mentre la configurazione al tempo $t_0-dt$ viene assegnata generando delle velocità random con centro di massa uguale a $0$.
- $1$ -> la configurazione al tempo $t_0$ viene letta dal file config.0, che contiene l'ultima configurazione della simulazione precedente. La configurazione al tempo $t_0 -dt$ viene letta dal file config.final che contiene la penultima configurazione della simulazione precedente.  
  
Quando il parametro restart è uguale a $1$ viene eseguito il procedimento sotto suggerito: 
- si caricano le configurazione al tempo $t_0$ e al tempo $t_0-dt$
- si esegue un *move* ottenendo la configurazione al tempo $t_0 +dt$ e si ricava la velocità come $\vec v(t_0) = \dfrac{\vec r (t_0+dt)-\vec r(t_0-dt)} {2 dt} $ e si ottiene tramite il teorema di equipartizione la temperatura attuale
- a questo punto si ricava il fattore di scala, per riscalare le velocità in modo tale da avere come temperatura iniziale la temperatura target a cui si vuole studiare il sistema fisico
- riscalate le velocità si ricava da queste la configurazione al tempo $t_0$

**input.dat**
- temperatura target (unità di LJ, $\frac{\epsilon}{k_B}$)
- numero di particelle ($108$)
- densità di particelle (unità LJ, $\frac{1}{\sigma^3}$)
- r_{cutoff} (unità LJ, $\sigma$)
- $\delta t$ (unità LJ, $\sigma \sqrt{\frac{m}{\epsilon}}$ che significa per l'Argon $10^{-15}sec$)
- numero di step per la simulazione
- ogni quanto stampare un cout di controllo per sapere a che punto sono

### Exercise 04.1

The Molecular Dynamics (MD) code starts the simulation run from a provided spatial configuration of the $N$ particles (file: <font face="Courier">config.0</font>) and, once the simulation is finished, it writes out the final configuration (file: <font face="Courier">config.final</font>). 
1. <span style="color:red">Improve the MD code by enabling (you should insert this option via a boolean-like variable in the input file) the **possibility** to **(re)start** reading not only a **actual** spatial configuration [$\vec{r}(t)$] file but also an **old** spatial configuration [$\vec{r}(t-dt)$] file</span> (files: <font face="Courier">old.0</font> and <font face="Courier">old.final</font>).
2. Moreover, <span style="color:red">add the possibility to rescale the velocities [i.e. to correct $\vec{r}(t-dt)$] in order to match a desired temperature</span>.
3. In your report, put some pictures which show that you have been able to equilibrate the thermodynamic phases at the beginning of your simulations.

The typical required operations when the re-start option will be enabled are:
1. read $\vec{r}(t)$ and $\vec{r}(t-dt)$ from the corresponding input files
2. compute $\vec{r}(t+dt)$ with one step of the Verlet algorithm; with $\vec{r}(t+dt)$ and $\vec{r}(t)$ compute $\vec{v}(t+dt/2)$; finally obtain the actual temperature $T(t+dt/2)$
3. by comparing $T(t+dt/2)$ with the desired/target temperature  $T^\star$ extract a scaling factor for the velocities and rescale them: $\vec{v}(t) \to \vec{v}_s(t)$
4. use $\vec{v}_s(t)$ to estimate a novel **old** spatial configuration: $\vec{r}_{new}(t) = \vec{r}(t+dt) - dt \vec{v}_s$
5. use $\vec{r}_{new}(t)$ and $\vec{r}(t+dt)$ to start the simulation
6. Do not forget to use **periodic boundary contitions** every time it is appropriate!

This will help you, especially in the equilibration phases of your simulations.

### Prima parte dei risultati
La prima parte della simulazione quindi consiste nel portare il sistema alla temperatura target.  
I sistemi studiati sono:  
- solido $(T^* = 0.8)$,$(\rho^* =1.1)$
- liquido $(T^* = 1.1)$,$(\rho^* =0.8)$
- gas $(T^* = 1.2)$,$(\rho^*=0.05)$  
  
Per comprendere perchè stiamo effettivamente parlando di queste fasi occorre ricorrere alla visualizzazione del diagramma delle fasi dei fludi di Lennard-Jones (per questo motivo ho posto anche il valore della densità)  
  
<img src="LJdiagram2.png" width="300">

Nei seguenti grafici mostro effettivamente come il sistema viene portato gradualmente alla temperatura assegnata.  
Si può notare come per ogni singola simulazione venga conservata l'energia. Questo è un buon segno della fatto che stiamo integrando in modo corretto e con un integratore buono. Il nostro sistema è descritto dall'insieme statistico microcanonico, $(N,V,E)$ per cui l'energia totale del sistema si deve conservare.  
Per ogni simulazione sono stati eseguiti 10'000 step MD, cioè mosse move() in cui vengono aggiornate la posizioni delle particelle in base alle forze agenti su di esse tramite l'algoritmo di Verlet: $$\vec r (t+dt)=2 \vec r(t)-\vec r(t-dt) + dt^2 \vec a(t)$$  
Le forze sono calcolate utilizzando $-\vec \nabla V$, dove $V$  il potenziale di Lennard-Jones $$V_{ij}^{LJ} = 4 \epsilon [ (\frac{\sigma}{|\vec r_i-\vec r_j|})^{12} - (\frac{\sigma}{|\vec r_i-\vec r_j|})^{6} ]$$  
Le quantità fisiche misurate sono:  
- temperatura $T^*$
- energia potenziale per particella $V^*$
- energia cinetica per particella $K^*$
- energia totale per particella $U^*$
- pressione $P^*$
- pair correlation function, radial correlation function $g(r^*)$  
  
Per evitare correlazioni tra le misure di queste quantità durante l'esecuzione del codice, quindi durante il ciclo su nstep, si calcolano queste quantità ogni $10$ step. Si possono calcolare anche ogni step, ma necessariamente saranno correlate tra di loro (occorrerebbe calcolare la lunghezza di correlazione) poichè l'algoritmo per integrare esprime chiaramente una dipendenza dalla configurazione attuale ma anche da quella precedente, correlando la futura correlazione anche a quest'ultima.  
Un'ulteriore cosa importante da tenere a mente è che sto simulando un box del nostro sistema, per cui è fondamentale utilizzare le condizioni periodiche al contorno e far interagire una particella anche con le "immagini" della particelle presenti nei box circostanti.  

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

nstep=10000 
fig,axs = plt.subplots(nrows=1,ncols=3,figsize=(15, 5))
folder = "raggiungimento_eq/solid/","raggiungimento_eq/liquid/","raggiungimento_eq/gas/"
name = "output_ekin"+str(nstep),"output_epot"+str(nstep),"output_temp"+str(nstep),"output_etot"+str(nstep),"output_press"+str(nstep)
color = "blue","red","orange","green","black"
properties = "kinetic energy","potential energy","temperature","total energy","pressure","target temperature"
title = "solid","liquid","gas"
temperature = 0.8,1.1,1.2


for j in range(3):
    for i in range(5):
        val = np.loadtxt(folder[j]+name[i]+".dat",usecols=(0),delimiter=' ',unpack='true')
        if (i == 4):
            axs[j].plot(val,color=color[i],alpha=0.7)
        else:
            axs[j].plot(val,color=color[i])
        axs[j].set_ylabel("physical properties")
        axs[j].set_xlabel("nstep")
    y = np.full(len(val),temperature[j])
    x = np.arange(len(val))
    axs[j].plot(x,y)
    axs[j].set_title(f"reaching target temperature of {temperature[j]} for "+title[j])
    axs[j].legend(properties)
    
plt.show()
```

### COMMENTO
Come ci si poteva aspettare ci vuole più tempo a far raggiungere la temperatura desiderata ai sistemi fisici liquido e gas, rispetto al solido. Questo perchè questi sistemi essendo molto più disordinati hanno maggiore libertà di movimento (le singole particelle) e quindi difficili da equilibrare (soprattutto il gas).    
Le simulazioni sono state eseguite una di seguito all'altra. Il solido partiva da una configurazione fcc. Il liquido dall'ultima configurazione del solido e il gas dall'ultima del liquido

```python
fig,axs = plt.subplots(nrows=1,ncols=3,figsize=(15, 5))
folder = "solid/","liquid/","gas/"
e_col = "red","blue","green","silver","cyan","ivory","orchid"

for j in range(3):
    for i in range(5):
        val = np.loadtxt(folder[j]+name[i]+".dat",usecols=(0),delimiter=' ',unpack='true')
        axs[j].plot(val,color=color[i])
        axs[j].set_ylabel("physical properties")
        axs[j].set_xlabel("nstep")
    y = np.full(len(val),temperature[j])
    x = np.arange(len(val))
    axs[j].plot(x,y)
    axs[j].set_title(f"simulation for "+title[j])
    axs[j].legend(properties)
    
fig,axs = plt.subplots(nrows=1,ncols=3,figsize=(15, 5))

temp = "ave_temp"+str(nstep)
properties = "target temperature","system temperature"

for j in range(3):
    sum_prog,err_prog = np.loadtxt(folder[j]+temp+".out",usecols=(0,1),delimiter=' ',unpack='true')
    x = np.arange(len(sum_prog))
    axs[j].errorbar(x,sum_prog,yerr=err_prog,color=color[j],ecolor=e_col[j])
    axs[j].set_ylabel("temperature")
    axs[j].set_xlabel("nblocks")
    y = np.full(len(sum_prog),temperature[j])
    axs[j].plot(x,y)
    axs[j].set_title(f"data blocking for "+title[j])
    axs[j].legend(properties)

plt.show()
```

```python
title = "kinetic_energy","potential_energy","temperature","total_energy","pressure"
name = "ave_ekin","ave_epot","ave_temp","ave_etot","ave_press"
name_properties = "output_ekin","output_epot","output_temp","output_etot","output_press"

nstep=100000
N=100
x = np.arange(N)

y_lab = "Kinetic En., [LJ]","Potential En., [LJ]","temperature, [LJ]","Total En., [LJ]","Pressure, [LJ]"

k=0
fig,axs = plt.subplots(nrows=3*3,ncols=2,figsize=(20, 70))
for p in range(3):
    k=0
    x = np.arange(N)
    for j in range(3):
            for i in range(2):
                if (j==2 and i==1):
                    y = np.loadtxt(folder[p]+"output.gave.out",usecols=(0),unpack='true')
                    sum_prog,err_prog = np.loadtxt(folder[p]+"output.gave.out",usecols=(1,2),unpack='true')
                    axs[j+3*p][i].errorbar(y,sum_prog,yerr=err_prog,color=e_col[1],ecolor=e_col[0])
                    axs[j+3*p][i].set_xlabel("r*",fontsize=20)
                    axs[j+3*p][i].set_ylabel("$g(r)$",fontsize=20)
                    axs[j+3*p][i].set_title("pair radial correlation function",fontsize=20)
                else:
                    sum_prog,err_prog = np.loadtxt(folder[p]+name[k]+str(nstep)+".out",usecols=(0,1),delimiter=' ',unpack='true')
                    axs[j+3*p][i].errorbar(x,sum_prog,yerr=err_prog,color=color[k],ecolor=e_col[1+p])
                    axs[j+3*p][i].set_xlabel("nblocks",fontsize=20)
                    axs[j+3*p][i].set_ylabel(y_lab[k],fontsize=30)
                if(p==0 and j==0 and i==0):
                    axs[j+3*p][i].set_title("solid Argon",fontsize=30)
                if(p==1 and j==0 and i==0):
                    axs[j+3*p][i].set_title("liquid Argon",fontsize=30)
                if(p==2 and j==0 and i==0):
                    axs[j+3*p][i].set_title("gas Argon",fontsize=30)
                k+=1

plt.show()
```

\frac{1}{2} \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^6 \right] \right\rangle $$
